```python
df_clean=df[(df['Sales']<lower_bound)& (df['Sales']>upper_bound)]
print("Orinigal rows:",len(df))
print("Rows after removing outliers:",len(df_clean))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Profit'], bins=30)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Sales'])

plt.title("Boxplot of Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Profit'])

plt.title("Boxplot of Profit")
plt.show()
```

```python
plt.figure(figsize=(10,5))

df.groupby('Category')['Sales'].sum().plot(kind='bar')

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
print(df.columns.tolist())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(12,5))
plt.plot(monthly_sales.index, monthly_sales.values)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

# Key Insights
1. Technology category generated the highest sales among all categories
2. Sales distribution is highly right-skewed, indicating a few very large orders
3. Profit contains both positive and negative values, showing that some orders resulted in losses
4. Sales vary significantly across different categories and sub-categories
5. Monthly sales generally increased over time, with noticeable seasonal fluctuations
